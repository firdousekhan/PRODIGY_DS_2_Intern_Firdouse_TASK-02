```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Train.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df.head(10)
```

```python
# Drop the Cabin column as it has too many missing values
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.head(10)
```

```python
# Fill missing values in the "Embarked" column with the most common value
most_common_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(most_common_embarked, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Visualize the count of survivors
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
# Visualize the count of survivors based on gender
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')
plt.show()
```

```python
# Visualize the count of survivors based on passenger class
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
# Visualize the distribution of age
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Visualize the survival rate based on age
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Survival by Age')
plt.show()
```

```python
# Visualize the count of passengers based on embarkation port
sns.countplot(x='Embarked', data=df)
plt.title('Passenger Count by Embarkation Port')
plt.show()
```

```python
# Visualize the count of survivors based on embarkation port
sns.countplot(x='Survived', hue='Embarked', data=df)
plt.title('Survival Count by Embarkation Port')
plt.show()
```
